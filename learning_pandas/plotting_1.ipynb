```python
import pandas as pd
```

```python
new_labels = ['series', 'series_code', 'country_name', 'country_code', 'total_population']
```

```python
df = pd.read_csv(
    'Health_Nutrition_and_Population_Statistics/Total_Population/2018.csv', header=0, names=new_labels, dtype={'total_population': float}, na_values='..'
)
```

```python
df = df.dropna(subset=['country_code', 'total_population'])
```

```python
df = df.sort_values(by='total_population')
```
